```python
from ultralytics import YOLO
```

### Training

```python
model = YOLO("yolo11n.pt")
```

```python
yaml_path = '/content/defectos_de_soldaduras/data.yaml'
```

#### Option 1: Default configuration

```python
model.train(
    data=yaml_path,
    epochs=10,
    imgsz=640,
    augment=True,
    )
```

#### Option 2: Patience with Early Stopping

If the model doesn't improve in 5 epoch with loss validation, stops the training

```python
model.train(
    data=yaml_path,
    epochs=20,
    imgsz=640,
    patience=5,
    augment=True,
    )
```

#### Option 3: Custom Data Augmentation

```python
model.train(
    data=yaml_path,
    epochs=10,
    imgsz=640,
    patience=5,
    augment=True,
    hsv_h=0.015, #Adjusting tone HSV
    hsv_s=0.7,   #Adjusting saturation HSV
    hsv_v=0.4,   #Adjusting brightness HSV
    flipud=0.5,  #Fliping vertical with 50% or probability
    fliplr=0.5,  #Fliping horizontal with 50% or probability
    mosaic=0.5,  #Turn on mosaic (images mixing)
    mixup=0.5,   #Mixing two images with 50% prbability
    perspective=0.0005, #Transformation of perspective
    degrees=5.0,   #Rotation of +- 5 grades
    scale=0.5,   #Zoom in the image
    shear=2.0,   #Inclination of the image
    )
```

#### Return especific results

```python
def detecting_objects(image_path, model):
  """
  Making the detection of objects on an image and returns:
    - The bounding boxes in format [x1,y1,x2,y2]
    - The names of the detected classes
    - The confindece of each detection
    - the marked image with detections

  Parametres:
    image_path (str): path to the image
    model (YOLO): model to use

  Returns:
    boxes (numpy.ndarray): bounding boxes
    class_names (list): names of the detected classes
    confidences (list): confidence of each detection
    annotated_img (numpy.ndarray): marked image with detections
  """

  #Making detection (to an image, results on an np.ndarray and take first result)
  results = model(image_path)
  result = results[0]

  #Extract the bounding boxes
  boxes = result.boxes.xyxy.cpu().numpy()

  #Extract indices of affected classes and confidences
  classes_idx = result.boxes.cls.cpu().numpy().astype(int)
  confidences = result.boxes.conf.cpu().numpy()

  #If the model has dictionary names, we obtain the names of the classes
  if hasattr(model, "names"):
    class_names = [model.names[i] for i in classes_idx]
  else:
    class_names = [str(i) for i in classes_idx] #If not exists, returns indices

  #Obtain the image with bounding boxes
  annotated_img = result.plot()

  return boxes, class_names, confidences, annotated_img
```

```python
image_path = "/content/defectos_de_soldaduras/valid/images/SampleV1_2_mp4-35_jpg.rf.112706bdf7b84566f9b8fa03084da685.jpg"

boxes, class_names, confidences, annotated_img = detecting_objects(image_path, my_model)
```
